```python
import numpy as np

df_dirty = df.copy()  # always work on a copy, never the original
df_dirty.loc[10:50, 'Sales'] = np.nan
df_dirty.loc[200:210, 'Region'] = np.nan

print("Missing values introduced:")
print(df_dirty.isnull().sum()[df_dirty.isnull().sum() > 0])
```

```python
#Drop rows with any missing value:
df_dropped = df_dirty.dropna()
print("Original rows:", len(df_dirty))
print("After dropna:", len(df_dropped))
print("Rows lost:", len(df_dirty) - len(df_dropped))
```

```python
# Fill missing numbers with the mean:
df_filled = df_dirty.copy()
mean_sales = df_filled['Sales'].mean()
df_filled['Sales'] = df_filled['Sales'].fillna(mean_sales)
print(f"Filled missing Sales with mean: ${round(mean_sales, 2)}")
print("Missing Sales remaining:", df_filled['Sales'].isnull().sum())
```

```python
#Fill missing text with the most common value:
most_common_region = df_filled['Region'].mode()[0]
df_filled['Region'] = df_filled['Region'].fillna(most_common_region)
print(f"Filled missing Region with: {most_common_region}")
print("Missing Region remaining:", df_filled['Region'].isnull().sum())
```

```python
#Fill missing values differently per column (real-world pattern):
fill_strategy = {
    'Sales': df_dirty['Sales'].median(),   # median for skewed numbers
    'Region': df_dirty['Region'].mode()[0] # mode for categories
}

df_clean = df_dirty.fillna(fill_strategy)
print("All missing values after smart fill:")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
print("Done â no missing values!" if df_clean.isnull().sum().sum() == 0 else "")
```

```python
#Create two separate tables to merge:
# Table 1: order-level data
orders = df[['Order ID', 'customer_name', 'Category', 'Sales', 'Profit']].copy()

# Table 2: customer segment info
customer_info = df[['customer_name', 'Segment', 'Region']]\
                .drop_duplicates(subset='customer_name')\
                .copy()

print("Orders table shape:", orders.shape)
print("Customer info shape:", customer_info.shape)
print("\nOrders sample:")
print(orders.head(3))
print("\nCustomer info sample:")
print(customer_info.head(3))
```

```python
 #Inner join (only matching rows):
merged_inner = orders.merge(customer_info, on='customer_name', how='inner')
print("Inner join shape:", merged_inner.shape)
print(merged_inner.head(3))
```

```python
# Left join (keep all orders, add customer info where available):
merged_left = orders.merge(customer_info, on='customer_name', how='left')
print("Left join shape:", merged_left.shape)
print("Missing customer info:", merged_left['Segment'].isnull().sum())
```

```python
# Create two tiny test dataframes to see exactly what each join does
df_a = pd.DataFrame({
    'id':    [1, 2, 3, 4],
    'value': ['A', 'B', 'C', 'D']
})

df_b = pd.DataFrame({
    'id':    [2, 3, 5],
    'extra': ['X', 'Y', 'Z']
})

print("Table A:"); print(df_a)
print("\nTable B:"); print(df_b)
print("\nINNER JOIN (only matching ids 2,3):")
print(df_a.merge(df_b, on='id', how='inner'))
print("\nLEFT JOIN (all of A, match where possible):")
print(df_a.merge(df_b, on='id', how='left'))
print("\nOUTER JOIN (everything from both):")
print(df_a.merge(df_b, on='id', how='outer'))
```

```python
merged = orders.merge(customer_info, on='customer_name', how='left')

segment_analysis = merged.groupby('Segment').agg(
    total_profit = ('Profit', 'sum'),
    avg_profit   = ('Profit', 'mean'),
    order_count  = ('Order ID', 'count')
).round(2)

print(segment_analysis)
```

## Key Findings

### 1. Discounting is destroying profit
- Orders with **no discount** average **$66.90 profit**
- Orders with **41-60% discount** average **-$134.62 loss**
- Discount and Profit have a **-0.22 correlation** â mathematically confirmed
- **Recommendation:** Cap discounts at 20%. Above that, the company loses money.

### 2. Furniture is a problem category
- Furniture profit margin: **2.49%**
- Technology margin: **17.40%**, Office Supplies: **17.04%**
- Furniture in the **Central region is loss-making** (-$2,871 total)
- **Recommendation:** Review Furniture pricing in Central region immediately.

### 3. Large orders are far more valuable
- Large orders (>$1,000) average **$314 profit**
- Small orders (<$100) average only **$5.35 profit**
- Large orders are 59x more profitable but only 5% of volume
- **Recommendation:** Sales team should prioritise and incentivise large order customers.

### 4. 3D printers should be removed from catalogue
- Cubify CubeX Double Head: **-$8,879 total loss**
- Cubify CubeX Triple Head: **-$3,839 total loss**
- No technology product should consistently lose this much
- **Recommendation:** Discontinue 3D printer product line.

### 5. Home Office segment is underserved
- Highest avg profit per order: **$33.82**
- Smallest segment by volume: only 1,783 orders
- **Recommendation:** Increase marketing focus on Home Office customers.
